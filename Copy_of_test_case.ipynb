```python
# Separate numerical and categorical data
numdata = data.select_dtypes(include=['int', 'float'])
catdata = data.select_dtypes(include='object')
```

```python
date_columns = ['last_date_of_month_6', 'last_date_of_month_7', 'last_date_of_month_8', 'last_date_of_month_9',
                'date_of_last_rech_6', 'date_of_last_rech_7', 'date_of_last_rech_8', 'date_of_last_rech_9',
                'date_of_last_rech_data_6', 'date_of_last_rech_data_7', 'date_of_last_rech_data_8',
                'date_of_last_rech_data_9']
for column in date_columns:
    data[column] = pd.to_datetime(data[column])
```

```python
# Extract day from date columns
day_data = data[date_columns].apply(lambda x: x.dt.day)
```

```python
# Drop categorical columns
data.drop(columns=catdata.columns, inplace=True)
```

```python

# Fill null values in numerical data with 0
numdata.fillna(0, inplace=True)
```

```python

# Concatenate numerical data and day data along columns axis
newdata = pd.concat([numdata, day_data], axis=1)
```

```python
# Calculate total revenue for each customer
newdata['total_revenue'] = (newdata['arpu_6'] + newdata['arpu_7'] +
                            newdata['arpu_3g_6'] + newdata['arpu_3g_7'] +
                            newdata['arpu_2g_6'] + newdata['arpu_2g_7'] )

# Calculate the 80th percentile of total revenue to identify high-value customers
high_value_threshold = newdata['total_revenue'].quantile(0.7)

# Filter high-value customers based on the threshold
high_value_customers = newdata[newdata['total_revenue'] >= high_value_threshold]

# Display high-value customers
print("High-Value Customers:")
print(high_value_customers)
```

```python
high_value_customers.isnull().sum()
```

```python
# Fill null values in date columns with zeros
date_data_columns = ['last_date_of_month_6', 'last_date_of_month_7', 'last_date_of_month_8',
       'last_date_of_month_9', 'date_of_last_rech_6', 'date_of_last_rech_7',
       'date_of_last_rech_8', 'date_of_last_rech_9',
       'date_of_last_rech_data_6', 'date_of_last_rech_data_7',
       'date_of_last_rech_data_8', 'date_of_last_rech_data_9']
high_value_customers[date_data_columns] = high_value_customers[date_data_columns].fillna(0)
```
